# Visualización de Datos
La Visualización de Datos es la práctica de convertir datos crudos a representaciones gráficas como mapas, gráficas, grafos y dashboards. Hace a un dataset complejo, más accesible, entendible y utilizable.

## Matplotlib
Para comprender mejor los datos, es útil poder visualizarlos de varias maneras. Matplotlib es la biblioteca de visualización más común para Python. Puede crear gráficos de líneas, diagramas de dispersión, diagramas de densidad, histogramas, mapas de calor, etc.

```python
import numpy as np
np.set_printoptions(suppress=True)
data_x = np.linspace(-100,100,100)
data_y = np.power(data_x, 2)
data_y
```

```python
#FORMA 1
plt.figure(figsize=(6,3))
plt.plot(data_x, data_y, color="green", label="al cuadrado")
plt.plot(data_x, data_x ** 3, label="al cubo")
plt.grid(True)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.show()
```

```python
%matplotlib inline
#FORMA 2: Orientada a objetos:
fig, axs = plt.subplots(figsize=(8,5), ncols=2, nrows=2)

axs[0][0].plot(data_x, data_y, color="green", label="al cuadrado")
axs[0][0].plot(data_x, data_x ** 3, label="al cubo")
axs[0][0].grid()
axs[0][0].set_xlabel("EJE X")
axs[0][0].set_ylabel("EJE Y")
axs[0][0].legend(loc="lower right")
axs[0][0].set_title("Título del axis 0 0")

axs[1][1].plot([100,200,300,400,500])

axs[0][1].plot(data_x, np.sin(data_x))

fig.tight_layout()

plt.show()
```

Al usar Matplotlib en un script de Python, las gráficas nos aparecerán en una ventana separada. Sin embargo, al trabajar con notebooks, podemos visualizar nuestros gráficos directamente en el documento (lo cual es ideal a la hora de generar informes con código y visualizaciones en un solo documento).

```python
%matplotlib widget
plt.figure(figsize=(6,3))
plt.plot(data_x, data_y, color="green", label="al cuadrado")
plt.plot(data_x, data_x ** 3, label="al cubo")
plt.grid(True)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend(loc="lower right")

plt.show()
```

### Gráfico de Dispersión(scatter)
La función scatter nos permite visualizar un conjunto de puntos.

```python

data_x_scatter = np.random.rand(10)
data_y_scatter = np.random.rand(10)
plt.scatter(data_x_scatter, data_y_scatter)
plt.show()

```

### Gráfico de Barras
Permite comparar de forma directa, valores cuantitativos por variable o categoría

```python
plt.bar(["2020", "2021", "2022"], [10000, 20000, 5000])
plt.title("VENTAS ANUALES")
plt.show()
```

### Histogramas
Un histograma es una representación de la estimación de la distribución de valores de un dataset numérico.

```python
data_normal = np.random.normal(0,1,10000)
plt.hist(data_normal, bins=200)
plt.show
```

### Gráfico de Torta
Permite analizar de forma básica de la composición porcentual de una `variable categórica`.

```python
data_pie = [400,400,200]
labels = ["gatos", "perros", "otros"]
plt.pie(data_pie, labels=labels, autopct="%1.1f%%")
plt.show()
```

### Gráfico de Caja y Bigotes
Este diagrama representa visualmente la distribución de los datos numéricos, así como la asimetría, mostrando los cuartiles y las medias de los datos. 

```python
plt.boxplot([18,60,30,51,29,30,500])
plt.show()
```
